```python
# ds = dataset
dsfile = '/content/traffic-volume-counts-1.csv'
ds = pd.read_csv(dsfile, delimiter='\t')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/traffic-volume-counts-1.csv'  # Update with your file path
data_new = pd.read_csv(file_path)

print("Dataset Info:")
print(ds.info())

# Filter numerical columns (hourly traffic counts)
hourly_data_new = data_new.iloc[:, 7:]  # Exclude the first few non-numerical columns

# Calculate correlation matrix
corr_new = hourly_data_new.corr()

# Plot the heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(corr_new, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Hourly Traffic Volume Counts')
plt.show()
```

**Many variable pairs present correlation close to zero. What does that mean?**

When we say many variable pairs present correlation close to zero, it means that there is no relationship between two variables.

**Why separate the genders?**


When we separate the genders, we can achieve a clearer visualization of the data results. Additionally, this approach allows for a more distinct comparison between the two primary subjects, which is clearer than combining them into a single heatmap.

**What variables have stronger correlation with brain size (MRI_Count)? Is that expected? Explain.**

Variables such as weight and height have stronger correlations with brain size (MRI_Count), with weight showing a correlation of 0.65 and height 0.58. This is expected as larger body size often correlates with larger organ sizes, including the brain.
